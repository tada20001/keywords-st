### 파일 오픈

```python
df = pd.read_excel('키워드분석용_rawdata.xlsx').fillna('')
df.columns
```

### 한글키워드 합치기

```python
string = str_keyword(df)
len(string)
```

### 카운팅 안할 단어 지정

```python
stop_words = ['및', '환경', '오염', '|', '및', '초', '성', '기술', '시스템', \
              '구조', '의', '중심으로', '관한', '대한', ':', '기반', '효과', \
              '개발', '에', '과', '연구', '적', '방법', '위한','변화', '미치는',\
              '활용', '-', '난', '프로', '미세먼지']
```

### 토큰 만들기

```python
tokens = tokenizer(string, stop_words)
len(tokens)
```

```python
tokens[:10]
```

### 그래프 그리기

```python
text = drawPlot_and_Count_commonWords(tokens, 50, "전체그래프.png", (12, 3))
```

```python
text
```

### 가장 많이 출현한 데이터 표로 나타내기

```python
most_words, fd_names = get_most_common_words(tokens, 30)
most_words
```

```python
fd_names
```

### 워드클라우드 표시

```python
draw_wordcloud(fd_names, '전체_미세먼지')
```
